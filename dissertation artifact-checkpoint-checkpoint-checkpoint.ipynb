```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from difflib import SequenceMatcher
from torchvision import transforms
import torchvision.models as models
import seaborn as sns

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, precision_recall_curve
```

```python
input_folder = r'C:\Users\owen\Desktop\Dissertation\green_area'
output_folder_resized = r'C:\Users\owen\Desktop\Dissertation\Resized_green_area'
past_folder_path = r'C:\Users\owen\Desktop\Dissertation\Past'
current_folder_path = r'C:\Users\owen\Desktop\Dissertation\Current'

colour_thresholds = {'green': ((15, 50, 50), (100, 255, 250))}
num_classes = 2 
target_size = (200,200)

#resize images
def resize_images(input_folder, output_folder, target_size):
    os.makedirs(output_folder, exist_ok=True)
    image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

    for image_file in image_files:
        image_path = os.path.join(input_folder, image_file)
        image = cv2.imread(image_path)

        if image is None:
            print(f"failed to load image/s {image_file}")
            continue
        resized_image = cv2.resize(image, target_size)

        output_path = os.path.join(output_folder, image_file)
        cv2.imwrite(output_path, resized_image)
#save images
def save_image(image, folder, filename):
    os.makedirs(folder, exist_ok=True)
    output_path = os.path.join(folder, filename)
    cv2.imwrite(output_path, image)

#load and preprocess images
def load_and_preprocess_images(folder_path, target_size):
    images = []
    labels = []

    #loop through images
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(folder_path, filename)
            image = cv2.imread(image_path)
            if image is not None:
                #resize image
                image = cv2.resize(image, target_size)
                images.append(image)

                #encode label
                label = 1 if folder_path.endswith("green_area") else 0
                labels.append(label)

    return np.array(images), np.array(labels)

def calculate_image_name_similarity(past_folder_path, current_folder_path):
    past_images = os.listdir(past_folder_path)
    current_images = os.listdir(current_folder_path)

    #dictionary to store similarity scores
    similarity_scores = {}
    for i, current_image in enumerate(current_images):
        if i < len(past_images):
            past_image = past_images[i]
            similarity = SequenceMatcher(None, current_image, past_image).ratio()
            similarity_scores[(current_image, past_image)] = similarity

    return similarity_scores
#label images based on green pixels
def calculate_green_pixel_count(image_path, colour_thresholds):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Could not read image {image_path}")
        return None
    #image to HSV colour space
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    #checks for green
    green_mask = cv2.inRange(hsv, colour_thresholds['green'][0], colour_thresholds['green'][1])
    #calculate green pixel count
    green_pixel_count = cv2.countNonZero(green_mask)
    return green_pixel_count

def label_image_by_green_pixels(image_path, colour_thresholds):
    #calculate green pixel count for the image
    green_pixel_count = calculate_green_pixel_count(image_path, colour_thresholds)
    if green_pixel_count is None:
        return None
    image = cv2.imread(image_path)
    height, width, _ = image.shape
    
    #calculate the percentage of green pixels
    green_percentage = (green_pixel_count / (height * width)) * 100
    
    if green_percentage > 25:
        dominant_color = 'green'
    else:
        dominant_color = 'non-green'
    return dominant_color



#label images in the folder based on green or not green
for filename in os.listdir(output_folder_resized):
    if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(output_folder_resized, filename)
            labeled_category = label_image_by_green_pixels(image_path, colour_thresholds)
            green_pixel_count = calculate_green_pixel_count(image_path, colour_thresholds)
            print(f"Image {filename} labeled as {labeled_category}. green pixel count: {green_pixel_count}")
    else:
            print("")
        

```

```python
class CNNModel(nn.Module):
    def __init__(self, num_classes):
        super(CNNModel, self).__init__()
        self.features = nn.Sequential(
            #input channels=3 (RGB), output channels=32, kernel size=3x3, padding=1
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            #ReLU activation function
            nn.ReLU(inplace=True),
            #max pooling layer with kernel size=2x2, stride=2
            nn.MaxPool2d(kernel_size=2, stride=2),
            #input channels=32, output channels=64, kernel size=3x3, padding=1
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            #ReLU activation function
            nn.ReLU(inplace=True),
            #max pooling layer with kernel size=2x2, stride=2
            nn.MaxPool2d(kernel_size=2, stride=2),
            #input channels=64, output channels=128, kernel size=3x3, padding=1
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            #ReLU activation function
            nn.ReLU(inplace=True),
             #max pooling layer with kernel size=2x2, stride=2
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.fc_input_size = 128 * (target_size[0] // 8) * (target_size[1] // 8)
        self.classifier = nn.Sequential(
            nn.Linear(self.fc_input_size, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

#load and preprocess images
def load_and_preprocess_images(folder_path, colour_thresholds, target_size):
    images = []
    labels = []
    
    # Loop through images in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(folder_path, filename)
            labeled_category = label_image_by_green_pixels(image_path, colour_thresholds)
            if labeled_category is not None:
                #Load image and resize
                image = cv2.imread(image_path)
                image = cv2.resize(image, target_size)
                images.append(image)
                
                #encode label
                label = 1 if labeled_category == 'green' else 0
                labels.append(label)
    
    return np.array(images), np.array(labels)





images, labels = load_and_preprocess_images(output_folder_resized, colour_thresholds, target_size)

#Changes images and labels to PyTorch tensors
images_tensor = torch.tensor(images, dtype=torch.float32).permute(0, 3, 1, 2)
labels_tensor = torch.tensor(labels, dtype=torch.long)

#training and testing sets with an 80/20 ratio
train_images, test_images, train_labels, test_labels = train_test_split(images_tensor, labels_tensor, test_size=0.2, random_state=42)
print("Training set size:", len(train_images))
print("Testing set size:", len(test_images))
model = CNNModel(num_classes)

#loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())


batch_size = 32
epochs = 10
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for i in range(0, len(train_images), batch_size):
        inputs = train_images[i:i+batch_size]
        labels = train_labels[i:i+batch_size]

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss / len(train_images)}")
```

```python
model.eval()
predicted_labels = []
true_labels = []
from sklearn.metrics import precision_recall_curve
with torch.no_grad():
    for images, labels in zip(test_images, test_labels):
        outputs = model(images.unsqueeze(0))  #batch dimension
        _, predicted = torch.max(outputs.data, 1)
        predicted_labels.append(predicted.item())
        true_labels.append(labels.item())

#lists to numpy arrays
predicted_labels = np.array(predicted_labels)
true_labels = np.array(true_labels)

#evaluation metrics
accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels)
recall = recall_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels)
conf_matrix = confusion_matrix(true_labels, predicted_labels)


print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", conf_matrix)
precision, recall, _ = precision_recall_curve(true_labels, predicted_labels)

plt.figure()
plt.plot(recall, precision, color='blue', lw=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
class CNNModel(nn.Module):
    def __init__(self, num_classes, similarity_scores, green_pixel_counts_current, green_pixel_counts_past):
        super(CNNModel, self).__init__()
        
        #initialize with required arguments
        self.similarity_scores = similarity_scores
        self.green_pixel_counts_current = green_pixel_counts_current
        self.green_pixel_counts_past = green_pixel_counts_past
        
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Linear(512 * 7 * 7 + 3, 1024), 
            nn.ReLU(inplace=True),
            nn.Linear(1024, 512),
            nn.ReLU(inplace=True),
            nn.Linear(512, num_classes),
        )

    def forward(self, x):
        #extract features from CNN layers
        x = self.features(x)
        x = torch.flatten(x, 1)
        
        #calculate average similarity score and average green pixel count
        avg_similarity_score = sum(score for (_, _), score in self.similarity_scores.items()) / len(self.similarity_scores)
        avg_green_pixel_count_current = sum(self.green_pixel_counts_current.values()) / len(self.green_pixel_counts_current)
        avg_green_pixel_count_past = sum(self.green_pixel_counts_past.values()) / len(self.green_pixel_counts_past)
        
        additional_features = torch.tensor([[avg_similarity_score, avg_green_pixel_count_current, avg_green_pixel_count_past]])  # Assuming batch size of 1
        additional_features = additional_features.repeat(x.size(0), 1) 
        x = torch.cat((x, additional_features), dim=1)
        
        x = self.classifier(x)
        return x

# Function to calculate similarity score between image names
def calculate_image_name_similarity(past_folder_path, current_folder_path):
    past_images = os.listdir(past_folder_path)
    current_images = os.listdir(current_folder_path)

    similarity_scores = {}

    for i, current_image in enumerate(current_images):
        if i < len(past_images):
            past_image = past_images[i]
            similarity = SequenceMatcher(None, current_image, past_image).ratio()
            similarity_scores[(current_image, past_image)] = similarity
    
    return similarity_scores

# Function to calculate green pixel count in an image
def calculate_green_pixel_count(image_path, colour_thresholds):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Could not read image {image_path}")
        return None

    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    green_mask = cv2.inRange(hsv, colour_thresholds['green'][0], colour_thresholds['green'][1])
    green_pixel_count = cv2.countNonZero(green_mask)
    
    return green_pixel_count

# Define folders and parameters
past_folder_path = r'C:\Users\owen\Desktop\Dissertation\Past'
current_folder_path = r'C:\Users\owen\Desktop\Dissertation\Current'
colour_thresholds = {'green': ((15, 50, 50), (100, 255, 255))}
output_folder_resized = r'C:\Users\owen\Desktop\Dissertation\Resized_green_area'  # Replace with your output folder
target_size = (224, 224)  # Replace with your desired target size

# Load and preprocess images
def load_and_preprocess_images(folder_path, colour_thresholds, target_size):
    images = []
    labels = []
    

    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(folder_path, filename)
            labeled_category = label_image_by_green_pixels(image_path, colour_thresholds)
            if labeled_category is not None:
                # Load image and resize
                image = cv2.imread(image_path)
                image = cv2.resize(image, target_size)
                images.append(image)
                
                #encode label
                label = 1 if labeled_category == 'green' else 0
                labels.append(label)
    
    return np.array(images), np.array(labels)

images, labels = load_and_preprocess_images(output_folder_resized, colour_thresholds, target_size)

#convert images and labels to PyTorch tensors
images_tensor = torch.tensor(images, dtype=torch.float32).permute(0, 3, 1, 2)
labels_tensor = torch.tensor(labels, dtype=torch.long)
train_images, test_images, train_labels, test_labels = train_test_split(images_tensor, labels_tensor, test_size=0.2, random_state=42)

print("Training set size:", len(train_images))
print("Testing set size:", len(test_images))

#calculate similarity score and display green pixel count
similarity_scores = calculate_image_name_similarity(past_folder_path, current_folder_path)

green_pixel_counts_current = {}
green_pixel_counts_past = {}

for (current_image, past_image), similarity in similarity_scores.items():
    current_image_path = os.path.join(current_folder_path, current_image)
    past_image_path = os.path.join(past_folder_path, past_image)

    green_pixel_count_current = calculate_green_pixel_count(current_image_path, colour_thresholds)
    green_pixel_count_past = calculate_green_pixel_count(past_image_path, colour_thresholds)
    
    green_pixel_counts_current[current_image] = green_pixel_count_current
    green_pixel_counts_past[past_image] = green_pixel_count_past

#create the CNN model with the computed scores and counts
model = CNNModel(num_classes, similarity_scores, green_pixel_counts_current, green_pixel_counts_past)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())

#train the model
batch_size = 32
epochs = 10 
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for i in range(0, len(train_images), batch_size):
        inputs = train_images[i:i+batch_size]
        labels = train_labels[i:i+batch_size]

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss / len(train_images)}")
```

Ensemble Model

```python
class EnsembleModel(nn.Module):
    def __init__(self, num_classes, similarity_scores, green_pixel_counts_current, green_pixel_counts_past):
        super(EnsembleModel, self).__init__()
        self.cnn_model = CNNModel(num_classes)
        #required arugments 
        self.similarity_scores = similarity_scores
        self.green_pixel_counts_current = green_pixel_counts_current
        self.green_pixel_counts_past = green_pixel_counts_past
        #linear layers for additional features
        self.linear_similarity = nn.Linear(1, 1)  #similarity score
        self.linear_current_count = nn.Linear(1, 1)  #current green pixel count
        self.linear_past_count = nn.Linear(1, 1)  #past green pixel count
        
        #classifier for the combined model
        self.Ensemble_classifier = nn.Sequential(
            nn.Linear(num_classes + 3, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_classes)
        )

    def forward(self, images):
        cnn_features = self.cnn_model.features(images)
        cnn_features = torch.flatten(cnn_features, 1)
        #average similarity score
        avg_similarity_score = sum(score for (_, _), score in self.similarity_scores.items()) / len(self.similarity_scores)
        avg_similarity_score = self.linear_similarity(torch.tensor([[avg_similarity_score]]))
        #average current green pixel count
        avg_green_pixel_count_current = sum(self.green_pixel_counts_current.values()) / len(self.green_pixel_counts_current)
        avg_green_pixel_count_current = self.linear_current_count(torch.tensor([[avg_green_pixel_count_current]]))
        
        #average past green pixel count
        avg_green_pixel_count_past = sum(self.green_pixel_counts_past.values()) / len(self.green_pixel_counts_past)
        avg_green_pixel_count_past = self.linear_past_count(torch.tensor([[avg_green_pixel_count_past]]))
        
        #concatenate additional features
        additional_features = torch.cat((avg_similarity_score, avg_green_pixel_count_current, avg_green_pixel_count_past), dim=1)
        
        #combine CNN with additional features
        ensemble_features = torch.cat((cnn_features, additional_features), dim=1)
        
        #forward pass through the combined classifier
        output = self.Ensemble_classifier(ensemble_features)
        
        return output
epochs = 10

train_losses = []

#training loop
for epoch in range(epochs):
    model.train()
    running_loss = 0.01
    for i in range(0, len(train_images), batch_size):
        inputs = train_images[i:i + batch_size]
        labels = train_labels[i:i + batch_size]

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
    epoch_loss = running_loss / len(train_images)
    train_losses.append(epoch_loss)
    print(f"Epoch {epoch + 1}/{epochs}, Loss: {epoch_loss}")
```
